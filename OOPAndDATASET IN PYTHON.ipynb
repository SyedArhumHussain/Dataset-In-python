```python
class Loc_Phone:
    def __init__(self):
        self.phone = None
    
    def input_phone(self):
        self.phone = input("Enter local phone number: ")
    
    def display_phone(self):
        print("Local phone number:", self.phone)


class Nat_Phone(Loc_Phone):
    def __init__(self):
        super().__init__()
        self.city_code = None
    
    def input_city_code(self):
        self.city_code = input("Enter city code: ")
    
    def display_city_code(self):
        print("City code:", self.city_code)


class In_Phone(Nat_Phone):
    def __init__(self):
        super().__init__()
        self.country_code = None
    
    def input_country_code(self):
        self.country_code = input("Enter country code: ")
    
    def display_country_code(self):
        print("Country code:", self.country_code)


phone1 = In_Phone()
phone1.input_country_code()
phone1.input_city_code()
phone1.input_phone()

phone1.display_country_code()
phone1.display_city_code()
phone1.display_phone()
```

```python
initial_count = len(df)
df.drop_duplicates(inplace=True)
duplicates_removed = initial_count - len(df)
print(f"\n{duplicates_removed} duplicates removed.")
print(df.head())

```

```python
missing_values = df.isnull().sum()
print("\nMissing values:")
print(missing_values)
print(df.head())


```

```python
df['Unit price'].fillna(df['Unit price'].mean(), inplace=True)
df['gross income'].fillna(df['gross income'].mode()[0], inplace=True)
print("\nMissing values filled.")
print(df.head())

```

```python
total_sales = df.groupby('Unit price')['gross income'].sum()
print("\nTotal sales of each product line:")
print(total_sales)
print(df.head())

```
